```python
import random

def generar_numero():
    """Genera un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)

def juego_adivinar_numero():
    numero_secreto = generar_numero()
    intentos = 0
    print("¡Bienvenido al juego de adivinar el número!")
    print("Estoy pensando en un número entre 1 y 100.")

    while True:
        try:
            intento_jugador = int(input("Adivina el número: "))
            intentos += 1

            if intento_jugador < numero_secreto:
                print("¡Demasiado bajo! Intenta de nuevo.")
            elif intento_jugador > numero_secreto:
                print("¡Demasiado alto! Intenta de nuevo.")
            else:
                print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
                break
        except ValueError:
            print("Entrada inválida. Por favor, ingresa un número entero.")

# Para iniciar el juego, puedes llamar a la función:
juego_adivinar_numero()
```

```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ValueError("No se puede dividir por cero.")
    return a / b

def calculadora():
    while True:
        print("\nSeleccione operación:")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Salir")

        opcion = input("Ingrese su opción (1/2/3/4/5): ")

        if opcion == '5':
            print("¡Gracias por usar la calculadora!\n")
            break

        if opcion in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Ingrese el primer número: "))
                num2 = float(input("Ingrese el segundo número: "))
            except ValueError:
                print("Entrada inválida. Por favor, ingrese números válidos.")
                continue

            if opcion == '1':
                print(f"{num1} + {num2} = {suma(num1, num2)}")
            elif opcion == '2':
                print(f"{num1} - {num2} = {resta(num1, num2)}")
            elif opcion == '3':
                print(f"{num1} * {num2} = {multiplicacion(num1, num2)}")
            elif opcion == '4':
                try:
                    print(f"{num1} / {num2} = {division(num1, num2)}")
                except ValueError as e:
                    print(e)
        else:
            print("Opción inválida. Por favor, intente de nuevo.")

# Para iniciar la calculadora, llama a la función:
calculadora()
```

```python
tareas = []

def agregar_tarea(tarea):
    tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea(indice):
    try:
        if 0 <= indice < len(tareas):
            tarea_eliminada = tareas.pop(indice)
            print(f"Tarea '{tarea_eliminada}' eliminada.")
        else:
            print("Índice de tarea inválido.")
    except TypeError:
        print("Entrada inválida. Por favor, ingrese un número para el índice.")

def mostrar_tareas():
    if not tareas:
        print("No hay tareas en la lista.")
    else:
        print("\n--- Lista de Tareas ---")
        for i, tarea in enumerate(tareas):
            print(f"{i}. {tarea}")
        print("-----------------------")

def lista_de_tareas():
    while True:
        print("\n--- Menú de Tareas ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Mostrar tareas")
        print("4. Salir")
        print("----------------------")

        opcion = input("Ingrese su opción: ")

        if opcion == '1':
            tarea = input("Ingrese la nueva tarea: ")
            agregar_tarea(tarea)
        elif opcion == '2':
            mostrar_tareas()
            try:
                indice = int(input("Ingrese el número de la tarea a eliminar: "))
                eliminar_tarea(indice)
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número entero.")
        elif opcion == '3':
            mostrar_tareas()
        elif opcion == '4':
            print("Saliendo del programa de lista de tareas. ¡Adiós!")
            break
        else:
            print("Opción inválida. Por favor, intente de nuevo.")

# Para iniciar la aplicación de lista de tareas, puedes llamar a la función:
lista_de_tareas()
```
