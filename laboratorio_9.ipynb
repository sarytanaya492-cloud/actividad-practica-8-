```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de ventas mensuales (en miles)
data = {"Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto",
"Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60,
70, 80, 75, 90]}

# Crear un DataFrame
df_ventas = pd.DataFrame(data)
display(df_ventas)
```

```python
# Usar operaciones de NumPy para el análisis
ventas = df_ventas['Ventas'].values # Convertir la columna de ventas a un array de NumPy

venta_promedio = np.mean(ventas)
venta_minima = np.min(ventas)
venta_maxima = np.max(ventas)
venta_std = np.std(ventas)

print(f"Venta promedio mensual: {venta_promedio:.2f} mil")
print(f"Venta mínima mensual: {venta_minima} mil")
print(f"Venta máxima mensual: {venta_maxima} mil")
print(f"Desviación estándar de ventas: {venta_std:.2f} mil")
```

```python
# Visualización de los resultados con Matplotlib
plt.figure(figsize=(12, 6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', linestyle='-', color='b')
plt.title('Ventas Mensuales (en miles)')
plt.xlabel('Mes')
plt.ylabel('Ventas (miles)')
plt.grid(True)
plt.xticks(rotation=45) # Rotar las etiquetas de los meses para mayor legibilidad
plt.tight_layout() # Ajustar el diseño para evitar el recorte de etiquetas
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes
data = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah",
"Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80,
88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]}

# Crear un DataFrame
df_calificaciones = pd.DataFrame(data)
display(df_calificaciones)
```

```python
# Usar operaciones de NumPy para el análisis de calificaciones

# Calificaciones promedio por asignatura
promedio_matematicas = np.mean(df_calificaciones['Matemáticas'])
promedio_ciencias = np.mean(df_calificaciones['Ciencias'])
promedio_historia = np.mean(df_calificaciones['Historia'])

print(f"Promedio de Matemáticas: {promedio_matematicas:.2f}")
print(f"Promedio de Ciencias: {promedio_ciencias:.2f}")
print(f"Promedio de Historia: {promedio_historia:.2f}")

print("\n--- Análisis por estudiante ---")
# Promedio de calificaciones por estudiante
df_calificaciones['Promedio_General'] = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean(axis=1)

display(df_calificaciones.sort_values(by='Promedio_General', ascending=False))
```

```python
# Visualización de los resultados con Matplotlib

# Gráfico de barras para calificaciones promedio por asignatura
plt.figure(figsize=(10, 6))
asignaturas = ['Matemáticas', 'Ciencias', 'Historia']
promedios = [np.mean(df_calificaciones['Matemáticas']), np.mean(df_calificaciones['Ciencias']), np.mean(df_calificaciones['Historia'])]
plt.bar(asignaturas, promedios, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Calificaciones Promedio por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gráfico de barras para el promedio general de cada estudiante
plt.figure(figsize=(12, 6))
plt.bar(df_calificaciones['Estudiante'], df_calificaciones['Promedio_General'], color='lightsalmon')
plt.title('Promedio General de Calificaciones por Estudiante')
plt.xlabel('Estudiante')
plt.ylabel('Promedio General')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
